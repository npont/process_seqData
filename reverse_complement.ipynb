# Reverse Complement

Before running sequana demultiplex, we need to reverse the sequence of index i5 (smart-seq). Therefore, provide the sample sheet here.

```python
from Bio.Seq import Seq
import pandas as pd
```

```python
# File in which we will put the reverse sequence (that will be used for Demultiplexing)
samplesheet_file="/sequencing_directory/SampleSheet.csv"
# File from which we take the sequence to reverse (that is directly provided by the sequencer or the biologist)
seq_info_file="/sequencing_directory/EXP<NUMBER>_Demultiplex_gDNA.xlsx"

samplesheet = pd.read_csv(samplesheet_file, header=None)
seq_info = pd.read_excel(seq_info_file)
```

```python
seq_info
```

```python
samplesheet
```

Reverse complement

```python
seq_info["reverse"] = seq_info["i5 sequence"].map(lambda x: str(Seq(x).reverse_complement()))
```

Replace i5 by reverse in samplesheet

```python
samplesheet[8] = seq_info["reverse"]
```

```python
samplesheet.to_csv("/sequencing_directory/SampleSheet.csv")
```
